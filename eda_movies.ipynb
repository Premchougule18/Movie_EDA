```python
numerc_cols=['BudgetUSD','US_BoxOfficeUSD','Global_BoxOfficeUSD','Opening_Day_SalesUSD','One_Week_SalesUSD','IMDbRating','RottenTomatoesScore']
```

```python
for col in numerc_cols:
    sns.histplot(df[col],bins=50,kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
    
```

```python
df.columns
```

```python
categorical_cols=['Genre','Country','Director','LeadActor']
```

```python
for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Top 20 {col}counts')
    plt.show()
    
    
```

```python
sns.scatterplot(x='BudgetUSD',y='Global_BoxOfficeUSD',data=df,alpha=0.3)
plt.title('Budget vs Global Box Office')
plt.xscale('log')
plt.yscale('log')   
plt.show()
```

```python
sns.scatterplot(x='IMDbRating',y='RottenTomatoesScore',data=df,alpha=0.3)
plt.title('IMDb Rating vs Rotten Tomatoes Score')
plt.show()
```

```python
df.columns
```

```python
genre_year_stats=df.groupby(['Genre','ReleaseYear']).agg({
    "Global_BoxOfficeUSD":'mean',
    "BudgetUSD":'mean',
    "IMDbRating":'mean',
    "RottenTomatoesScore":'mean'
})
```

```python
genre_year_stats
```

```python
plt.figure(figsize=(25,12))
sns.lineplot(data=genre_year_stats.reset_index(),x='ReleaseYear',y='Global_BoxOfficeUSD',hue='Genre')
plt.title('Average Global Box Office by Genre Over Time')
plt.show()
```

## Visual insights
The following charts compare budget, box office performance, genre trends, and actor ratings to uncover patterns in the data.

```python
top_actors=df.groupby('LeadActor')["IMDbRating"].mean().sort_values(ascending=False).head(20)

```

```python
top_actors
```

```python
top_actors.plot(kind='bar')
plt.title('Top 20 Lead Actors by Average IMDb Rating')
plt.show()
```
